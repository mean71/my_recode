## 깃 프로그램

- CLI(Command line interface): Gitbash
  - 터미널 입력으로 git 사용 -> Linux 명령어 및 git 명령어를 익혀야 사용가능
  - 대부분의 개발자들은 이 방식으로 깃을 사용
- GitHub Desktop: GUI로 구현 쉽게 사용가능 -> github기본기능 제공: 어딘가 나사가 조금 빠져 있음
- GitKraken: 기능성이 아주 좋은 GUI로 쉽고 편리하게 확인가능 -> 비공개저장소는 유료
- Source Tree(소스트리): 무료 - 깃의 기본 ~ 고급기능까지 준수한 gui를 제공: 사용법복잡, 익숙해지면 자유롭게 깃을 활용가능
- TortoiseGit(토터스깃) : 윈도우 전용 -> 윈도우탐색기의 빠른 메뉴에 추가되는 프로그램
- 더 많은 깃 프로그램 및 설치: -> https://git-scm.com/downloads/guis

- VSCode extention: vscode의 확장으로 별도의 툴없이 git과 github의 기능을 한번에 활용가능
  - gitgraph: 해당 깃 레포지토리의 변경기록을 그래프로 시각화
  - githistory: 코드와 문서에서 우측에 변경사항, 사용자, 시간등 기록을 표시 외 편의기능
  - github pull requests: pull requests를 vscode내에서 편하게 요청
  - GitLens - Git supercharged: 깃크라켄에서 제작한 시각화확장툴, 유료지만, 무료로 제공되는 그래프 시각화기능만으로 충분

- 개인적으로 vscode & 소스트리 조합을 사용, 깃크라켄은 시각화가 잘 되어있어 무료레포지토리 작업확인에 유용

## Git bash

- Git Bach화면에 다음 명령어 입력
0. git
1. git --help
2. (최초1회)사용자 정보 등록 : 이름 & 이메일 -> 누가 변경했는지 커밋기록이 (이름 이메일)로 구분되어 표시, 수정하기 귀찮
   1. git config --global user.name "이름"
   2. git config --global user.email "이메일@mail.com"
3. git init : (프로젝트진행할 폴더에서 Git시작, 해당폴더의 모든 진행과정이 저장되는 폴더(.git)생성)
4. git status : 저장소의 상태 / working tree status 확인

5. staging: 변경사항을 커밋하기전에 stage에 올리는 것 -> 변경사항 선택취소가능
    - git add 파일.명 # 해당 파일을 staging area로 추가
    - git add ./폴더명/ # 해당 폴더를 staging area로 추가
    - git add . #현재 디렉토리에 있는 모든 파일을  staging area로 추가
    - git add -A : (프로젝트폴더 모든 내부파일을 깃의 대상에 포함?)
6. commit: 현재 staging 되어 있는 모든 변경내용을 분기점으로 commit 하는것
    - git commit -m "메세지" # 특정메세지를 작성하여 버전생성
      - 캡쳐하듯 현재 작업내용을 깃에 저장. 프로젝트에 의미있는 분기점,변화마다 '커밋'을 통해 작업내용보관
      - 전체가 아닌 이전 버전과의 변경사항만 기록.
    - git commit -am "메세지" # git add와 commit을 한번에 가능(한번이상 커밋한 파일만 가능)
    - git commit # 여러줄의 커밋 메세지를 편집기를 통해 편리하게 작성
    - git commit --amend #이전 커밋메세지를 편집기로 수정
7. git log 확인: 해당 디렉토리의 git 변경 기록을 탐색
    - git logs #로컬저장소의 커밋히스토리를 탐색(q버튼으로 종료)
    - git log --stat # 버전마다 관련파일들을 그룹핑하여 확인
    - git log -p # 아주 복잡한 코드일 경우 문제코드 추적시 확인하기 좋다
8. git push: 로컬에서 commit된 최종 분기점을 연결된 원격저장소로 push(업로드)
    - git diff #변경사항 확인가능 커밋전에 확인하면 좋다.
    - push:
9. git분기점 복원
    - git reset --hard 6자리커밋ID # 현재 작업내용을 삭제하고 해당 분기점으로 복원(공유 이전 단계의 버전만 가능?)
    - git reset --hard HEAD^ # 최종 커밋취소하고 파일까지 복구한다는데 분기점을 다 지워버린다는건가?
    - git revert 현재버전커밋ID # 기존커밋은 놔두고 이전커밋으로 복원. reset과 달리 기존커밋이 보존된다. 4번째에서 1번째로 가려면 3번실행하는식
10. git branch: main이 되는 분기점에서 분리진행되는 분기 생성
    - git branch "branch이름"
      - 실험적 시도, 신기능 시험시 기존 f프로젝트와 분리진행가능. 주 프로젝트에 영향없이 여러기능 구현된 여러버전생성가능.
11. git merge: 서로 다른 branch를 하나로 병합
    - git merge "branch이름"

			PDF4 GitHub로 백업(미완?완성?학습미완료)
1. 원격저장소
-지역저장소(local repository)
   작업을 수행한 후 커밋을 저장한 컴퓨터
-원격저장소(remote repository)
   지역 저장소가 아닌 컴퓨터나 서버에 만든 저장소
   지역 저장소와 연결되어 있으면서 백업 및 협업을 가능하게 함
   인터넷에서 원격 저장소를 제공하는 깃허브 주로 사용
   지역저장소생성>git init>원격저장소생성>(등록)>git remote add>git add>git commit>git push
2.  GitHub
- 깃관련 가장 많이 사용되는 원격저장소 제공서비스
- 특징
   무료,유료서비스, 오픈소스들
   git설치없이도 온라인 버전관리가능
   지역저장소 온라인백업가능
   협업에 사용

# 커밋 convention
1. 정의
형상관리 툴인 Git을 사용하면서 Commit을 할 때 Commit Message를 작성하는 규칙
2. 커밋 메세지 형식
정의하기 나름이지만, 보통 Udacity Style을 많이 사용
type: subject, body, footer로 나뉘고 각 파트는 한 줄의 공백을 두어 구분

커밋 메시지 
타입 : 
feat: 새로운 기능 추가
fix: 버그 수정
docs: 문서 수정
style: 코드 스타일 변경 (코드 포맷팅, 세미콜론 누락, 코드 변경이 없는 경우)
refactor: 기능 변경 없이 코드 리팩토링
test: 테스트 코드 추가,리펙토링 테스트 코드 추가
chore: 빌드 및 유지 관리 작업(빌드 업무 수정, 패키지 매니저 수정)
comment: 주석 추가 및 변경
remove: 파일 또는 폴더 삭제
rename: 파일 또는 폴더 이름 변경
주제
커밋에 대한 짧은 설명 (최대 50글자)
본문
변경 사항에 대한 자세한 설명 (최대 72글자)
바닥글
이슈 트래커 ID를 포함하고, 수정 또는 해결에 대한 설명.
예시 참조
Fixes: 아직 해결되지 않은 이슈
Resolves: 해결된 이슈
Ref: 관련 이슈
Related to: 아직 해결되지 않은 이슈

이 형식을 통해 커밋 이력을 명확히 하고 팀원 간의 협업을 쉽게 합니다.
