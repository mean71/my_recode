### Vim 기본 명령어 및 개념 정리

---

#### 1. 모드 개념

1. Normal 모드: 기본 모드 (명령 실행)
   - `Esc`로 진입
2. Insert 모드: 텍스트 입력 모드
   - `i` (커서 앞), `I` (줄의 처음)
   - `a` (커서 뒤), `A` (줄의 끝)
   - `o` (아래 새 줄), `O` (위에 새 줄)
3. Visual 모드: 텍스트 블록 선택 모드
   - `v` (문자 단위), `V` (줄 단위), `Ctrl+v` (블록 선택)
4. Command-line 모드: 명령어 입력 모드
   - `:` 명령어 입력 (예: `:w`, `:q`, `:wq`)

---

#### 2. 파일 관리

- `:w` → 저장
- `:q` → 종료
- `:wq` → 저장 후 종료
- `:q!` → 저장하지 않고 강제 종료
- `:e [파일명]` → 파일 열기
- `:w [새파일명]` → 다른 이름으로 저장

---

#### 3. 이동 명령어

- `h` → 왼쪽으로 한 글자
- `l` → 오른쪽으로 한 글자
- `j` → 아래로 한 줄
- `k` → 위로 한 줄
- `0` → 줄의 맨 앞으로
- `^` → 줄의 첫 번째 글자로
- `$` → 줄의 맨 끝으로
- `gg` → 파일 맨 위로
- `G` → 파일 맨 아래로
- `:[숫자]` → 특정 줄로 이동 (예: `:10`)

---

#### 4. 편집 명령어

- `x` → 커서 위치 글자 삭제
- `dd` → 현재 줄 삭제
- `d$` → 현재 위치부터 줄 끝까지 삭제
- `dw` → 단어 삭제
- `u` → 실행 취소
- `Ctrl+r` → 실행 취소 복구
- `yy` → 현재 줄 복사
- `p` → 붙여넣기 (복사한 내용 아래)
- `P` → 붙여넣기 (복사한 내용 위)
- `r` + `[문자]` → 한 글자 대체

---

#### 5. 검색 & 치환

- `/[검색어]` → 앞으로 검색
- `?[검색어]` → 뒤로 검색
- `n` → 같은 방향으로 다음 결과
- `N` → 반대 방향으로 다음 결과
- `:%s/old/new/g` → 파일 전체에서 `old`를 `new`로 변경
- `:s/old/new/` → 현재 줄에서 `old`를 `new`로 변경

---

#### 6. 화면 분할 (윈도우 관리)

- `:split` 또는 `:sp [파일명]` → 수평 분할
- `:vsplit` 또는 `:vsp [파일명]` → 수직 분할
- `Ctrl+w, w` → 분할된 창 전환
- `Ctrl+w, c` → 현재 창 닫기
- `Ctrl+w, q` → 창 종료

---

#### 7. 탭 관리

- `:tabnew [파일명]` → 새 탭 열기
- `gt` → 다음 탭으로 이동
- `gT` → 이전 탭으로 이동
- `:tabclose` → 현재 탭 닫기

---

#### 8. 매크로 사용

- `q[a-z]` → 매크로 녹화 시작 (예: `qa`는 a 레지스터에 기록)
- `[명령어 입력]`
- `q` → 녹화 종료
- `@[a-z]` → 매크로 실행 (예: `@a`는 a 레지스터 매크로 실행)
- `@@` → 마지막 매크로 재실행

---

#### 9. 자주 사용하는 VSCode Vim 단축키

- `Ctrl + [` → Esc와 동일 (Normal 모드로 전환)
- `Ctrl + d` → 화면 아래로 스크롤
- `Ctrl + u` → 화면 위로 스크롤
- `Ctrl + w, h/j/k/l` → 창 전환

---

#### 10. Vim 설정 파일 (~/.vimrc or VSCode settings.json)

- 기본 설정:
  ```
  set number         " 줄 번호 표시
  set relativenumber " 상대 줄 번호 표시
  set tabstop=4      " 탭 간격 설정
  set shiftwidth=4   " 들여쓰기 설정
  set expandtab      " 탭 대신 공백 사용
  syntax on          " 문법 하이라이팅
  ```

---

#### 11. 크롬 확장 Vimium 단축키

- `j/k` → 스크롤 아래/위로
- `h/l` → 좌/우로 이동
- `gg/G` → 페이지 최상단/최하단 이동
- `/` → 페이지 내 검색
- `f` → 모든 링크에 단축키 표시 (링크 열기)
- `F` → 새 탭으로 링크 열기
- `d/u` → 반 페이지 아래/위로 이동

---

#### 12. 자주 발생하는 오류 및 해결법

- 명령어 입력 시 알 수 없는 에러 발생  
  → `Esc` 누른 뒤 다시 시도 (모드 확인)
  
- 붙여넣기 시 자동 들여쓰기 오류  
  → `:set paste`로 붙여넣기 모드 활성화 후 붙여넣기, 완료 후 `:set nopaste`로 복귀

- VSCode에서 Vim 플러그인 반응 없음  
  → 플러그인 재설치하거나 `settings.json`에서 충돌 키맵 확인
