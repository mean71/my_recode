## CMD, PowerCell

- CMD: DOS 명령어 기반 전통 CLI: (기본적인 파일 및 디렉토리 관리, 시스템 정보 확인, 프로그램 실행 등 간단한 명령어들이 주로 사용)
  - DOS 명령어 기반 (텍스트 출력 기반 파이프라인)
  - 간단한 파일 및 디렉토리 작업, 시스템 작업, 프로그램 실행
  - 스크립트 기능: 간단한 배치(.bat) 파일
- PowerShell: 객체지향 스크립트 자동화 툴: 편리한 powershell ISE GUI 지원원
  - cmdlet명령어 기반, 객체 기반, .NET Framework 기반 (객체 기반 파이프라인)
  - cmd의 명령어와 기능을 대부분 지원하지만 몇가지 예외는 "cmd /c 명령어"로 실행 하거나 start-Process 명령어를 통해 외부 프로그램이나 cmd명령어 실행가능
  - 파이프라인, 스크립트, 함수, 변수 등을 사용하여 더 복잡한 시스템 작업과 자동화 가능
  - 스크립트 기능: 고급 스크립트(.ps1) 및 함수, 변수 지원
  - 시스템관리, 네트워크 관리, active directory 관리, 보안 설정 등까지 포함

PowerShell 명령어 총집합
1. 파일 및 디렉토리 작업
  - 디렉토리 이동
    - CMD: `cd [디렉토리 경로]`
    - PowerShell: `cd [디렉토리 경로]` 또는 `Set-Location [디렉토리 경로]`
  - 현재 디렉토리 보기
    - CMD: `dir`
    - PowerShell: `Get-ChildItem` 또는 `ls`
  - 디렉토리 생성
    - CMD: `mkdir [디렉토리 이름]`
    - PowerShell: `New-Item -ItemType Directory -Name [디렉토리 이름]`
  - 파일 생성
    - CMD: `echo > [파일 이름]`
    - PowerShell: `New-Item -ItemType File -Name [파일 이름]`
  - 파일 삭제
    - CMD: `del [파일 이름]`
    - PowerShell: `Remove-Item [파일 이름]`
  - 파일 내용 보기
    - CMD: `type [파일 이름]`
    - PowerShell: `Get-Content [파일 이름]` 또는 `cat [파일 이름]`
  - 파일 복사
    - CMD: `copy [원본 파일 경로] [목적지 파일 경로]`
    - PowerShell: `Copy-Item [원본 파일 경로] -Destination [목적지 경로]`
  - 파일 이동
    - CMD: `move [원본 파일 경로] [목적지 경로]`
    - PowerShell: `Move-Item [원본 파일 경로] -Destination [목적지 경로]`
  - 파일 이름 변경
    - CMD: `ren [기존 파일 이름] [새 파일 이름]`
    - PowerShell: `Rename-Item [기존 파일 이름] -NewName [새 파일 이름]`
  - 폴더 및 파일 속성 보기
    - CMD: `dir /a`
    - PowerShell: `Get-ItemProperty [파일/폴더 경로]`

2. 시스템 정보 및 작업 관리
  - 시스템 정보 보기
    - CMD: `systeminfo`
    - PowerShell: `Get-ComputerInfo`
  - 프로세스 목록 보기
    - CMD: `tasklist`
    - PowerShell: `Get-Process` 또는 `ps`
  - 특정 프로세스 종료
    - CMD: `taskkill /PID [프로세스 ID]`
    - PowerShell: `Stop-Process -Name [프로세스 이름]` 또는 `Stop-Process -Id [프로세스 ID]`
  - 네트워크 정보 보기
    - CMD: `ipconfig`
    - PowerShell: `Get-NetIPAddress`
  - 컴퓨터 재부팅
    - CMD: `shutdown /r`
    - PowerShell: `Restart-Computer`
  - 컴퓨터 종료
    - CMD: `shutdown /s`
    - PowerShell: `Stop-Computer`

3. 파일 및 디렉토리 검색
  - 파일 검색
    - CMD: `dir [파일 이름] /s`
    - PowerShell: `Get-ChildItem -Recurse -Filter [파일 이름]`
  - 내용 검색
    - CMD: `findstr [문자열] [파일 이름]`
    - PowerShell: `Select-String -Path [파일 경로] -Pattern [문자열]`

4. 시스템 관리 및 스크립트 실행
  - PowerShell 스크립트 실행
    - CMD: `powershell -File [스크립트 경로]`
    - PowerShell: `.\[스크립트 이름]`
  - 배치 파일 실행
    - CMD: `[배치 파일 경로]`
    - PowerShell: `Start-Process [배치 파일 경로]`
  - 실행 정책 확인
    - CMD: 해당 명령어 없음
    - PowerShell: `Get-ExecutionPolicy`
  - 실행 정책 변경
    - CMD: 해당 명령어 없음
    - PowerShell: `Set-ExecutionPolicy [정책]`

5. 사용자 및 권한 관리
  - 사용자 정보 보기
    - CMD: `net user`
    - PowerShell: `Get-LocalUser`
  - 새 사용자 추가
    - CMD: `net user [사용자 이름] [비밀번호] /add`
    - PowerShell: `New-LocalUser -Name [사용자 이름] -Password (ConvertTo-SecureString "[비밀번호]" -AsPlainText -Force)`
  - 사용자 삭제
    - CMD: `net user [사용자 이름] /delete`
    - PowerShell: `Remove-LocalUser -Name [사용자 이름]`

6. 기타 PowerShell 전용 명령어
  - 변수 설정
    - CMD: `set [변수 이름]=[값]`
    - PowerShell: `$[변수 이름] = [값]`
  - 변수 출력
    - CMD: `echo %변수 이름%`
    - PowerShell: `$[변수 이름]`
  - 조건문
    - CMD: `if [조건] ...`
    - PowerShell: `if ([조건]) { ... }`
  - 반복문
    - CMD: `for %i in ([목록]) do ...`
    - PowerShell: `foreach ($item in [목록]) { ... }`
  - 파이프라인 사용
    - CMD: `명령어1 | 명령어2`
    - PowerShell: `명령어1 | 명령어2`
  - 함수 정의
    - CMD: 해당 명령어 없음
    - PowerShell:
      ```powershell
      function MyFunction {
          [명령어들]
      }
      ```
  - 날짜 및 시간 표시
    - CMD: `date` / `time`
    - PowerShell: `Get-Date`

7. 스크립트 및 자동화
  - 스크립트 실행
    - CMD: 배치 파일 실행
    - PowerShell: `Start-Process [스크립트 경로]`
  - 자동화 (작업 스케줄러 설정)
    - CMD: `schtasks`
    - PowerShell: `New-ScheduledTask` / `Register-ScheduledTask`

## CLI Command

1. 파일 및 디렉토리 작업
   - 디렉토리 이동: `cd [디렉토리 경로]`
   - 현재 디렉토리 보기: `pwd`
   - 디렉토리 생성: `mkdir [디렉토리 이름]`
   - 파일 생성: `touch [파일 이름]`
   - 파일 삭제: `rm [파일 이름]`
   - 디렉토리 삭제: `rm -r [디렉토리 이름]`
   - 파일 내용 보기: `cat [파일 이름]`
   - 파일 내용 추가: `echo "내용" >> [파일 이름]`
   - 파일 복사: `cp [원본 파일] [목적지 경로]`
   - 파일 이동: `mv [원본 파일] [목적지 경로]`
   - 파일 이름 변경: `mv [기존 파일 이름] [새 파일 이름]`
   - 파일 내용 검색: `grep [검색어] [파일 이름]`
   - 파일 및 디렉토리 목록 보기: `ls : ls -l : ls -a`
   - 파일 및 디렉토리 속성 보기: `stat [파일/디렉토리 이름]`

2. 시스템 관리 및 정보
   - 시스템 정보 보기: `uname -a : hostname : uptime`
   - CPU 정보 보기: `lscpu`
   - 메모리 사용량 보기: `free -h`
   - 디스크 사용량 보기: `df -h`
   - 프로세스 목록 보기: `ps : ps aux : top`
   - 프로세스 종료: `kill [프로세스 ID] : killall [프로세스 이름]`
   - 시스템 로그 보기: `dmesg : tail -f /var/log/syslog`

3. 사용자 및 권한 관리
   - 현재 사용자 정보 보기: `whoami`
   - 사용자 목록 보기: `cat /etc/passwd`
   - 새 사용자 추가: `sudo useradd [사용자 이름]`
   - 사용자 삭제: `sudo userdel [사용자 이름]`
   - 사용자 비밀번호 변경: `passwd [사용자 이름]`
   - 그룹 목록 보기: `cat /etc/group`
   - 사용자 그룹 변경: `sudo usermod -aG [그룹 이름] [사용자 이름]`

4. 네트워크 관리
   - 네트워크 인터페이스 보기: `ifconfig : ip addr`
   - 네트워크 연결 확인: `ping [IP 주소 또는 도메인]`
   - 네트워크 트래픽 보기: `netstat -tuln`
   - 포트 확인: `ss -tuln : lsof -i :[포트 번호]`
   - IP 주소 확인: `hostname -I`
   - 서버 정보 확인: `ssh [사용자명]@[IP 주소]`

5. 패키지 관리
   - Ubuntu/Debian 패키지 설치: `sudo apt-get install [패키지 이름]`
   - Ubuntu/Debian 패키지 제거: `sudo apt-get remove [패키지 이름]`
   - CentOS/RHEL 패키지 설치: `sudo yum install [패키지 이름]`
   - CentOS/RHEL 패키지 제거: `sudo yum remove [패키지 이름]`
   - 패키지 목록 업데이트: `sudo apt-get update : sudo yum update`
   - 설치된 패키지 목록 보기: `dpkg -l : rpm -qa`

6. 디스크 및 파일 시스템 관리
   - 디스크 파티션 확인: `fdisk -l`
   - 디스크 마운트: `mount [디스크 경로] [마운트 지점]`
   - 디스크 언마운트: `umount [마운트 지점]`
   - 디스크 용량 확인: `du -sh [디렉토리] : df -h`

7. 파일 압축 및 해제
   - 압축 파일 생성 (tar): `tar -cvf [압축파일명.tar] [디렉토리 또는 파일]`
   - 압축 파일 해제 (tar): `tar -xvf [압축파일명.tar]`
   - 압축 파일 생성 (gzip): `gzip [파일 이름]`
   - 압축 파일 해제 (gzip): `gunzip [파일 이름.gz]`
   - 압축 파일 생성 (zip): `zip [압축파일명.zip] [파일/디렉토리]`
   - 압축 파일 해제 (zip): `unzip [압축파일명.zip]`

8. 파일 및 디렉토리 권한 관리
   - 파일 권한 변경: `chmod [옵션] [파일 이름]`
   - 파일 소유자 변경: `chown [소유자] [파일 이름]`
   - 파일 그룹 변경: `chgrp [그룹 이름] [파일 이름]`
   - 권한 보기: `ls -l [파일/디렉토리 이름]`

9. 기타 유용한 명령어
   - 파일 내용 필터링: `grep [검색어] [파일 이름]`
   - 내용 출력: `less [파일 이름]`
   - 명령어 기록 보기: `history`
   - 명령어 찾기: `which [명령어]`
   - 시스템 재부팅: `sudo reboot`
   - 시스템 종료: `sudo shutdown -h now`
